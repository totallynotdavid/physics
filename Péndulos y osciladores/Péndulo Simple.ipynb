```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# Datos de entrada
y = float(input("Posición inicial: "))
z = float(input("Velocidad inicial: "))
```

```python
# Valores de las constantes y detalles
m = 1
l = 1
g = 9.81
h = 0.1
tf=10
t=0
n=int((tf-t)/h)
Y=[y]
Z=[z]
T=[t]
```

```python
def f(y,z):
    return -(g/l)*y
```

```python
# Método numérico
for i in range(n):
    k1=h*z
    d1=h*f(y,z)
    k2=h*(z+d1/2)
    d2=h*f(y+k1/2,z+d1/2)
    k3=h*(z+d2/2)
    d3=h*f(y+k2/2,z+d2/2)
    k4=h*(z+d3)
    d4=h*f(y+k3,z+d3)
    y = y+(k1+2*k2+2*k3+k4)/6
    z = z+(d1+2*d2+2*d3+d4)/6
    t = t+h
    Y.append(y)
    Z.append(z)
    T.append(t)
```

```python
# Gráfica
%matplotlib inline
plt.plot(T,Y)
plt.show()
```

```python
# Animación
%matplotlib auto
fig=plt.figure()
ax=fig.gca()
ax.set_facecolor('k')
def actualizar(i):
    ax.clear()
    plt.plot([0,l*np.sin(Y[i])],[0,-l*np.cos(Y[i])],'r')
    plt.plot(l*np.sin(Y[i]),-l*np.cos(Y[i]),'wo')
    plt.xlim(-l,l)
    plt.ylim(-l,0)
ani=animation.FuncAnimation(fig,actualizar,range(len(T)),interval=100)
plt.show()
```
